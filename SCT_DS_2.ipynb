### Step 1: Import necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = r"C:\Users\Lenovo\Downloads\Titanic-Dataset.csv"  
df = pd.read_csv(file_path)
```

### Step 2: Inspect dataset

```python
print("Describe data:")
print(df.describe)

print("First 5 rows of the dataset:")
print(df.head())

print("\nDataset Summary:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())
```

### Step 3: Data Cleaning

```python
# Fill missing values in 'Age' with median age
df['Age'].fillna(df['Age'].median(), inplace=True)

# Fill missing values in 'Embarked' with the most frequent value
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Drop 'Cabin' due to excessive missing values
df.drop(columns=['Cabin'], inplace=True)

# Convert categorical features into numerical format
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df['Embarked'] = df['Embarked'].map({'C': 0, 'Q': 1, 'S': 2})
```

### Step 4: Exploratory Data Analysis (EDA)

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', data=df, palette="coolwarm")
plt.title("Survival Count (0 = No, 1 = Yes)")
plt.show()

# Check survival rate by gender
plt.figure(figsize=(6, 4))
sns.barplot(x="Sex", y="Survived", data=df, palette="viridis")
plt.xticks(ticks=[0, 1], labels=["Male", "Female"])
plt.title("Survival Rate by Gender")
plt.show()

# Check survival rate by passenger class
plt.figure(figsize=(6, 4))
sns.barplot(x="Pclass", y="Survived", data=df, palette="Blues")
plt.title("Survival Rate by Passenger Class")
plt.show()
```

### Step 5: Correlation Heatmap

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
